Logistic Regression
===============

This is the demo for Logistic Regression




```python
data = np.loadtxt("Logistic_Regression.txt", delimiter=',')
colors = colors = ['red' if i == 0 else 'green' for i in data[:, 2]]

plt.scatter(data[:, 0], data[:, 1], c=colors)
plt.show()
```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
    return 1.0/(1+np.exp(-z))

def pad(x):
    return np.hstack((np.ones((x.shape[0], 1), dtype = x.dtype), x))

def update(theta, alpha, grad):
    return theta - alpha * grad

def h(theta, x):
    #YOUR CODE HERE
    return sigmoid(np.dot(x,theta))

def compute_gradient(theta, x, y):
    #YOUR CODE HERE
    n = x.shape[0]
    gradient = 1/n * np.dot(x.T, h(theta,x) - y)
#     gradient = 1/n * np.dot(x.T, y-h(theta,x))

    return gradient

def predict(theta, x):
    #YOUR CODE HERE
    return h(x, theta) > 0.5

x = pad(data[:, 0:2])
y = data[:, 2]
alpha = 2e-3
print(alpha)
tolerance = 1E-5
error = 1

theta = np.array([0.0, 0.0, 0.0])
iteration = 0;

while error > tolerance:
    grad = compute_gradient(theta, x, y)
    
    theta = update(theta, alpha, grad)
    
    error = alpha * np.amax(np.absolute(grad))
    iteration = iteration + 1
#     print(grad)
#     print(error)
    if iteration % 100000 == 0:
        print("Error ", error)

# theta = [-15.80041149,0.13148784,0.12574769]
print(theta)
```

```python
def compute_y(theta, x):
    #YOUR CODE HERE
    return np.dot(x,theta[0:2])/theta[2]
print(theta[0:2])
line_x = np.array([30, 100]).reshape([2,1])
line_y = -compute_y(theta, pad(line_x))
plt.scatter(data[:, 0], data[:, 1], c=colors)
plt.plot(line_x, line_y, '-')

plt.xlabel('number of cigarettes')
plt.ylabel('number of cups of alcohol')

plt.show()
```
